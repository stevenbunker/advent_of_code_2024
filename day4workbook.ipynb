```python
import numpy as np
import pandas as pd
```

```python
day = 4
input_file = f"inputs/day{day}input.text"
with open(input_file, 'r') as file:   

    characters = [list(char) for char in file.read().splitlines()]
    data = pd.DataFrame(characters)

data
```

```python
##solution 1
target = ['XMAS','SAMX']
total = 0
hvalue_count = dvalue_count = vvalue_count = odvalue_count = 0
i = 0
j = 0
rows = len(data)
columns = len(data.columns)
for row in range(rows):
    i = row
    for column in range(columns):
        j = column
        vvalue = ''.join(data[j][i:i+4])
        hvalue = ''.join(data.iloc[[j],i:i+4].agg(''.join, axis=1))
        dvalue = ''.join(pd.Series(np.diagonal(data.iloc[i:i+4,j:j+4])))
        odvalue = ''.join(pd.Series(np.diagonal(np.fliplr(data.iloc[i:i+4,j:j+4]))))
        if hvalue in target:
            hvalue_count += 1
        if vvalue in target:
            vvalue_count += 1
        if dvalue in target:
            dvalue_count +=1
        if odvalue in target:
            odvalue_count +=1
```

```python
##Solution 2
target = ['MAS','SAM']
total = 0
hvalue_count = dvalue_count = vvalue_count = odvalue_count = 0
i = 1
j = 1
rows = len(data)
columns = len(data.columns)
for row in range(1,rows-1):
    i = row
    for column in range(1,columns-1):
        j = column
        candidate = data.iloc[i,j]
        dvalue, odvalue = None, None
        if candidate == 'A':
            matrix = data.loc[i-1:i+1,j-1:j+1]
            dvalue = ''.join(pd.Series(np.diagonal(matrix)))
            odvalue = ''.join(pd.Series(np.diagonal(np.fliplr(matrix))))
        if set([dvalue,odvalue]) <= set(target):
            total += 1
total
```
